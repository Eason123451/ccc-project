```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
crime_data = pd.read_json('viccrime_by_offence_count.json')
population_data = pd.read_json('vicpopu.json')
```

```python
# Get a list of column names for population columns
population_cols = [col for col in population_data.columns if 'tpop_' in col and int(col.split('_')[1]) in range(2019, 2024)]

# Calculate the sum of population for each year and store in a list
population_sums = []
for col in population_cols:
    year = int(col.split('_')[1])
    pop_sum = population_data[col].sum()
    population_sums.append((year, pop_sum))

# Sort the list by year
population_sums.sort(key=lambda x: x[0])

# Print the results
for year, pop_sum in population_sums:
    print(f"Total population in {year}: {pop_sum}")
```

```python
population_df = pd.DataFrame(population_sums, columns=['year', 'population'])
merged_df = pd.merge(crime_data, population_df, left_on='filters', right_on='year', how='left')
merged_df = merged_df.drop(['filters','year'], axis=1)
```
